```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import imdb
```

```python
# Load IMDB dataset (top 10,000 most frequent words)
vocab_size = 10000

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

print("Training samples:", len(x_train))
print("Test samples:", len(x_test))

```

```python
# Reviews are sequences of word indices → need fixed length
max_length = 200

# Pad sequences (make all reviews same length)
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_length)
x_test = keras.preprocessing.sequence.pad_sequences(x_test, maxlen=max_length)
```

```python
model = keras.Sequential([

    # Convert word indices to dense vectors
    layers.Embedding(input_dim=vocab_size, output_dim=32, input_length=max_length),

    # Flatten to feed into dense layers
    layers.Flatten(),

    # Hidden layer
    layers.Dense(64, activation='relu'),

    # Output layer (binary classification)
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',   # Important for classification
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=128,
    validation_split=0.2
)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)

print("\nTest Accuracy:", accuracy)
```

```python
predictions = model.predict(x_test[:5])

for i in range(5):
    print(f"Prediction: {predictions[i][0]:.2f}, Actual: {y_test[i]}")
```
